```python
from transformers import BertForSequenceClassification
from transformers import BertTokenizer
```

```python
import torch
import numpy as np
```

```python
label_dict = {0: 'not sexist', 1 :'sexist'}
```

```python
#load tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased',
                                         do_lower_case = True)
```

```python
model_1 = "C:/Users/mufdu/Desktop/New folder/Pre-processed/Bert-based-uncased/model_A"
model = torch.load(model_1)
model.eval()
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
print(device)
```

```python
# from transformers import BertTokenizer, BertModel
# tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
# model = BertModel.from_pretrained("bert-base-uncased")
text = "Women can not do anything"
encoded_input = tokenizer(text, return_tensors='pt').to(device)
output = model(**encoded_input)
```

```python
print(output[0])
```

```python
logits = output[0].detach().cpu().numpy()
predictions = np.concatenate(logits, axis=0)
preds_flat = np.argmax(predictions, axis=0).flatten()

```

```python
print(label_dict[preds_flat[0]])
```
